```python
from gensim.models.fasttext import load_facebook_vectors

path = "cc.pl.300.bin"

model = load_facebook_vectors(path)
```

```python
def get_similar_words(word, topn=10):
    try:
        similar = model.most_similar(word, topn=topn)
        return [w for w, score in similar]
    except KeyError:
        return []
```

```python
import re

def group_similar_words_word2vec(text, threshold=0.7):

    words = re.findall(r'\b[a-zA-ZąćęłńóśźżĄĆĘŁŃÓŚŹŻ]+\b', text.lower())
    words_in_text = set(words)

    grouped = {}
    seen = set()

    for w in words_in_text:
        if w in seen or w not in model:
            continue

        similar = [
            (word, score)
            for word, score in model.most_similar(w, topn=50)
            if word in words_in_text and score > threshold
        ]

        if similar:
            group = {word for word, _ in similar}
            group.discard(w)

            if group:
                grouped[w] = sorted(group)
                seen.update(group)
                seen.add(w)

    return grouped
```

```python
with open("tekst.txt", "r", encoding="utf-8") as f:
    tekst = f.read()

wynik = group_similar_words_word2vec(tekst)

for rep, group in wynik.items():
    print(f"{rep}: {group}")

with open("results.txt", "w", encoding="utf-8") as out:
    for rep, group in wynik.items():
        out.write(f"{rep}: {', '.join(group)}\n")
```
